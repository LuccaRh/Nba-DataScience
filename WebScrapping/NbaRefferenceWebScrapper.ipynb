```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
dados_jogador = []
cabecalhos = []
url = 'https://www.basketball-reference.com/leagues/NBA_2024_per_game.html'
```

```python
response = requests.get(url)

if response.status_code == 200:
    
    soup = BeautifulSoup(response.text, 'html.parser') #Transforma html em beatifulsoup objeto, que faz representá-lo como DOM
    tabela = soup.find('table', {'id': 'per_game_stats'}) #Pega a tabela com os jogadores
    #Pega cabeçalho da tabela (thead), e dela, os nomes de todas suas colunas (th)  
    cabecalhos = [th.text.strip() for th in tabela.find('thead').find_all('th')] 
    
    #Loop para conteúdo da tabela (tbody), vendo cada linha/jogador (tr), pegando seus colunas/dados (td), e add na lista
    for linha in tabela.find('tbody').find_all('tr'):
        dados_jogador = [td.text.strip() for td in linha.find_all('td')]
        jogador = dict(zip(cabecalhos, dados_jogador))
        dados_jogadores.append(jogador)

else:
    print('Falha na solicitação HTTP:', response.status_code)
```

```python
df = pd.DataFrame(dados_jogadores)
df.head()
```
